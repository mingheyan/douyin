```python
data['100w粉丝时间节点']
formatted_date_list = [
    pd.to_datetime(date).strftime('%Y%m%d') if pd.notna(date) else None
    for date in data['100w粉丝时间节点']
]

# 将格式化后的日期列表赋值回DataFrame
data['时间节点'] = formatted_date_list
data.drop(data.index[73], axis=0, inplace=True)

```

```python
data
```

```python
# 将'时间节点'列中的字符串转换为datetime对象
# data['时间节点'] = pd.to_datetime(data['时间节点'])

# 在'时间节点'列的基础上加上45天
data['时间节点+20天'] = data['100w粉丝时间节点'] + pd.DateOffset(20)

# 将新的日期列格式化为'YYYYMMDD'格式的字符串
data['时间节点+20天'] = data['时间节点+20天'].dt.strftime('%Y%m%d')
# 在'时间节点'列的基础上加上45天
data['时间节点-20天'] = data['100w粉丝时间节点'] + pd.DateOffset(-20)

# 将新的日期列格式化为'YYYYMMDD'格式的字符串
data['时间节点-20天'] = data['时间节点-20天'].dt.strftime('%Y%m%d')
data
```

```python
data.to_csv('yan1.csv', index=False)
```

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta

# 获取当前时间
now = datetime.now()

# 获取当前时间的前一个月
one_month_ago = now - relativedelta(months=1)

# 格式化时间为'YYYYMMDD'形式的字符串
formatted_one_month_ago = one_month_ago.strftime('%Y%m%d')

print(formatted_one_month_ago)
```
